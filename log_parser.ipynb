```python
import matplotlib.pyplot as plt
import numpy as np
```

## Experiment: Original

```python
folder = 'original/run_2'

def parse_file(filename):
    file = open(filename, 'r')
    processed_log = []
    for idx, line in enumerate(file):
        if idx > 0:
            line_processed = line.split(',')
            line_processed[0] = int(line_processed[0])
            line_processed[1] = float(line_processed[1])
            line_processed[-1] = int(line_processed[-1])
            processed_log.append(line_processed)
        else:
            clk_freq = int(line[-2:])
    
    return clk_freq, np.array(processed_log)
def save(foldername, number):
    plt.savefig(foldername + '/{}.png'.format(number), dpi=144)

log_0_freq, log_0_results = parse_file(folder + '/log_original0.txt')
log_1_freq, log_1_results = parse_file(folder + '/log_original1.txt')
log_2_freq, log_2_results = parse_file(folder + '/log_original2.txt')
```

```python
plt.figure()

log_0_logical_clk_jumps = log_0_results[1:,0] - log_0_results[:-1,0]
log_1_logical_clk_jumps = log_1_results[1:,0] - log_1_results[:-1,0]
log_2_logical_clk_jumps = log_2_results[1:,0] - log_2_results[:-1,0]

plt.plot(range(len(log_0_logical_clk_jumps)), log_0_logical_clk_jumps, label = 'CLK FREQ {}'.format(log_0_freq))
plt.plot(range(len(log_1_logical_clk_jumps)), log_1_logical_clk_jumps, label = 'CLK FREQ {}'.format(log_1_freq))
plt.plot(range(len(log_2_logical_clk_jumps)), log_2_logical_clk_jumps, label = 'CLK FREQ {}'.format(log_2_freq))
plt.xlabel("Index")
plt.ylabel("Logical Clock Jump")

plt.legend()
plt.gcf().set_size_inches(5, 5)
save(folder, 1)
plt.show()
```

```python
plt.figure()

plt.plot(log_0_results[:,1], log_0_results[:,0], label = 'CLK FREQ {}'.format(log_0_freq))
plt.plot(log_1_results[:,1], log_1_results[:,0], label = 'CLK FREQ {}'.format(log_1_freq))
plt.plot(log_2_results[:,1], log_2_results[:,0], label = 'CLK FREQ {}'.format(log_2_freq))
plt.xlabel("System Timestamp")
plt.ylabel("Logical Clock")
plt.legend()
plt.gcf().set_size_inches(5, 5)
save(folder,2)
plt.show()
```

```python
plt.figure()

plt.plot(log_0_results[:,1], log_0_results[:,2], label = 'CLK FREQ {}'.format(log_0_freq))
plt.plot(log_1_results[:,1], log_1_results[:,2], label = 'CLK FREQ {}'.format(log_1_freq))
plt.plot(log_2_results[:,1], log_2_results[:,2], label = 'CLK FREQ {}'.format(log_2_freq))
plt.xlabel("System Timestamp")
plt.ylabel("Length of Message Queue")
plt.legend()
plt.gcf().set_size_inches(5, 5)
save(folder,3)
plt.show()
```
